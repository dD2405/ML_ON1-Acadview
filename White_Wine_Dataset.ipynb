# Data description

### White wine is mainly from "white" grapes, which are green or yellow in colour, such as the Chardonnay, Sauvignon, and Riesling. Some white wine is also made from grapes with coloured skin, provided that the obtained wort is not stained.
### Data are collected on 12 different properties of the wines one of which is Quality, based on sensory data, and the rest are on chemical properties of the wines including density, acidity, alcohol content etc. All chemical properties of wines are continuous variables. Quality is an ordinal variable with possible ranking from 1 (worst) to 10 (best). Each variety of wine is tasted by three independent tasters and the final rank assigned is the median rank given by the tasters.

It doesn't deal with causes or relationships (unlike regression) and it's major purpose is to describe; it takes data, summarizes that data and finds patterns in the data.The key pointers to the Univaraite analysis are to find out the outliers present in the data. We also tend to find the dsitribution of the data on the dataset which can further help us for the Bivaraite/Multivariate analysis.


# Bivariate Analysis

### Bivariate analysis is the simultaneous analysis of two variables (attributes). It explores the concept of relationship between two variables, whether there exists an association and the strength of this association, or whether there are differences between two variables and the significance of these differences.

## Co-relation 

```python
df.corr()
```

## Co-relation heatmap

```python
plt.figure(figsize =(18,18))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

## Alcohol vs Quality

### Here we will analyse the alcohol content and quality of a Wine

### These are the following inferences:-

1. Alcohol and Quality have co-relation 0.44

2. There is an increasing trend in the alcohol content as the quality increases

3. We can infer that the average quality wines rated 5 and 6 have alcohol content in the range 8.5 to 11.5 for rating 5 and alcohol content in the range 8.5 to 13 for rating 6

4. Quality rating 7 has most number of wines with alcohol content more than 11

```python
sns.regplot(x = 'quality',y = 'alcohol',data = df , x_jitter=0.2, scatter_kws={'alpha':0.1})
sns.despine()
```

## Quality vs pH 

### Here we analyse the Quality of wine and the acidity of wine by it's pH

### The following trends are observed:-

1. There is a slight increasing trend observed form the following plot
2. Wine quality rating 6 has maximum number of wines with pH of 3.2 or more
3. Quality rating 3 has lesser pH as compared to other qualities
4. quality and pH have a co-relation of 0.099


```python
sns.regplot(x = 'quality',y = 'pH',data = df , x_jitter=0.2, scatter_kws={'alpha':0.1})
sns.despine()
```

## Chlorides vs pH

### Here we compare the pH of wines with the chlorides content

# Multi-Variate Analysis

### Multivariate analysis (MVA) is based on the statistical principle of multivariate statistics, which involves observation and analysis of more than one statistical outcome variable at a time.

### Inferences form Alcohol vs Density in terms of Quality plot.

#### From the graphs below we can notice that higher quality rating wines have more alochol but it does not affect the density majorly.

```python
grid = sns.FacetGrid(df, col='quality',col_wrap = 2)
grid.map(plt.scatter,'alcohol','density',alpha = 0.1)
sns.despine()
```

### Inferences form Alcohol vs pH in terms of Quality plot.

#### From the graphs below we can infer that mediocre quality rated wines i.e. 5 and 6 have high amount of chlorides present in them as comapred to the high quality rated wines.Quality rating wines of 7 and 8 have high concentration of alcohol present in them i.e. more than 11%

```python
grid = sns.FacetGrid(df, col='quality',col_wrap = 2)
grid.map(plt.scatter,'alcohol','chlorides',alpha = 0.1)
sns.despine()
```

### Inferences form Alcohol vs residual.sugar in terms of Quality plot.

#### From the below graphs we can conclude that wines rated 5 and 6 have maximum number of samples. We also observe that wines having quality rating i.e 7 or 8 have high level of alcoholic concentration present.

```python
grid = sns.FacetGrid(df, col='quality',col_wrap = 2)
grid.map(plt.scatter,'alcohol','residual.sugar',alpha = 0.2)
sns.despine()
```
