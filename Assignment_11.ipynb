### Q-6. Perform Pre_Processing like Import Pandas library and Csv

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')
df.head(10)
```

### Q-4. Remove NaN values from 'Survived'

#### No null values so no removal as well

```python
df.isnull().sum()
```

### Q-5.Does Passenger have cabin or not?

```python
def has_cabin(col):
    if pd.isnull(col):
        return 0
    else:
        return 1
```

```python
df['Has_Cabin'] = df['Cabin'].apply(has_cabin)
df.head()
```

### Q-9 . Input Missing Values

```python
df.mean()
df=df.fillna(df.mean())
df.head(10)
```

### Q-10. Categorise age or fare for people

```python
df["Categorical_Age"]=pd.qcut(df["Age"], 5,labels=False)
df["Categorical_Fare"]=pd.qcut(df["Fare"], 7,labels=False)
df.head()
```

### Q-12. Transform variables into numeric

```python
# We change the value of the sex column categorise into male and female survival
dum = pd.get_dummies(df['Sex'])

# Now again we concatenate the dummy table with the original table 
df1 =  pd.concat([df,dum],axis=1)

df1.head()
```

### Q-11. Drop Columns

```python
df2 = df1.drop(columns=['Cabin', 'Name', 'PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch'])
df2.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
reg = LogisticRegression()
```

### Q-13. Split the dataset without 'Survived'

```python
labels = df2['Survived']
train1 = df2.drop(columns=['Sex','Embarked','Survived','Categorical_Age','Categorical_Fare'])
train1.head()
```

### Q-14. Pass your X-Train and Y_Train

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
reg.fit(x_train,y_train)
```

### Q-15 Print Accuracy Score

```python
reg.score(x_test,y_test)
```
