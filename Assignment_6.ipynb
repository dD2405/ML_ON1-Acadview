```
# Q1.
# a.)


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


#1
df1 = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')
df = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv')
df.dropna()
plt.title('Age and Fare')
sns.barplot(x="Embarked", y="Fare",data=df)
sns.despine()
```
