```
#Question 1

import numpy as np

a = np.random.rand(10,1)
a= a.mean()
print('Mean:',a)
```

```
#Question 2

import numpy as np

b = np.random.rand(20,1)
a = b.var()
c = b.std()

print('Variance:',a)
print('Standard Deviation:',c)
```

```
#Question 3

import numpy as np

a = np.random.rand(10,20)
b = np.random.rand(20,25)

c = np.dot(a,b)
e = c.reshape(10,25)
print(e)
d = sum(sum(c))
print('------------------------------------------\n')
print('Total Sum:',d)
```
