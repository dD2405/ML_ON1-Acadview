```
import numpy as np
import pandas as pd
```

```
data = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/house-price-prediction/master/kc_house_data.csv')
data.head()
```

```
from sklearn.linear_model import LinearRegression

# We call the function required from the package
reg = LinearRegression()

# We set this for testing of our dataset
labels = data['price']

# We here group the data for the column date by giving value 1 or 0
conv_dates = [1 if values == 2014 else 0 for values in data.date ]

data['date'] = conv_dates

# drop the id and price columns,as id is not required for predicting prices and 
# we have to predict the prices so we drop this column as well
train1 = data.drop(columns=['price','id'])
```

```
train1.head()
```

```
from sklearn.cross_validation import train_test_split
```

```
# We call the required function from the package for training and testing split
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.30,random_state =2)

# here we train the data 
reg.fit(x_train,y_train)
```

```
# Here we test our data and find the accuracy 
reg.score(x_test,y_test)
```
